## 2. Descrição do Problema e da Solução Proposta

<h3>O Índice de Desempenho da Saúde Suplementar (IDSS) é uma medida crucial que
reflete a qualidade e eficiência das operadoras de saúde. A análise tradicional
desses indicadores, porém, pode ser lenta e sujeita a erros. Além disso, muitas
vezes resulta em ações reativas, ao invés de proativas. A necessidade é de uma
solução que permita uma análise mais rápida, precisa e que possa ser utilizada para
tomar medidas preventivas e de aperfeiçoamento antes da manifestação de
problemas significativos. A previsão precisa do IDSS futuro tornou-se uma
necessidade imperativa, pois permite uma melhor planificação e tomada de decisão
tanto para as autoridades reguladoras quanto para as próprias operadoras.
Entretanto, essa previsão é complexa e desafiadora, devido à dinâmica 
multifacetada do setor de saúde e às constantes mudanças nas variáveis que
afetam o IDSS.

Para solucionar o problema da previsão do IDSS futuro, propomos a utilização de
técnicas avançadas de aprendizado de máquina. Este modelo não só poderá
fornecer previsões precisas para cada um dos indicadores, como também permitirá
que as operadoras e a ANS tomem decisões informadas e proativas. Dessa forma,
qualquer desvio no desempenho esperado poderá ser identificado com
antecedência, permitindo que sejam tomadas ações corretivas em tempo hábil.

O IDSS é um indicador que mede o desempenho das operadoras de planos de
saúde em diferentes aspectos ao longo do tempo. Esses indicadores podem incluir
qualidade, eficiência, satisfação do cliente, entre outros. Como os valores do IDSS
estão disponíveis para anos anteriores, podemos considerar essa situação como um
problema de previsão de séries temporais, onde o objetivo é prever os valores
futuros do IDSS com base nos valores históricos.

Os modelos de séries temporais são particularmente adequados para lidar com esse
tipo de problema, pois são projetados para capturar padrões temporais e tendências
presentes nos dados. Eles podem identificar sazonalidade, flutuações regulares ao
longo do tempo, bem como mudanças de longo prazo.

No caso do problema do IDSS, a sazonalidade pode ser influenciada por fatores
como épocas do ano, eventos regulatórios, mudanças nas práticas de gestão de
operadoras de saúde, entre outros. Os modelos de séries temporais, como o
<b>ARIMA, LSTM, SARIMA e a Regressão Linear</b>, são capazes de capturar essas
variações sazonais e incorporá-las nas previsões.

A solução proposta não apenas permitirá previsões mais precisas do IDSS futuro,
mas também fornecerá insights valiosos sobre os fatores que influenciam essas
mudanças. Isso permitirá às operadoras e reguladores tomar decisões mais
informadas, otimizando recursos e melhorando a qualidade dos serviços de saúde.

## 3. Coleta de Dados

<h3>Os dados estão disponibilizados de maneira pública, no Portal de Dados Abertos, do
governo.Intitulado como “Histórico do IDSS” podem ser acessados através do link:
https://dados.gov.br/dados/conjuntos-dados/historico-do-idss, com a última alteração:

os anos 2009 a 2021, são os seguintes indicadores:

<b>IDSS:</b> Índice de Desempenho da Saúde Suplementar, um indicador geral da
qualidade da operadora.

<b>IDQS:</b> Índice de Qualidade em Atenção à Saúde, que mede a qualidade dos
serviços de saúde prestados.

<b>IDSB:</b> Índice de Desempenho da Estrutura de Operações, que avalia a qualidade da
infraestrutura da operadora.

<b>IDSM:</b> Índice de Desempenho da Satisfação do Beneficiário, que mede a satisfação
do cliente com a operadora.

<b>IDEO:</b> Índice de Desempenho Econômico-Financeiro, que mede a estabilidade
financeira da operadora.

<b>IDGR:</b> Índice de Desempenho da Garantia de Atendimento, que mede a capacidade
da operadora de atender aos pedidos de serviço dos clientes.

<b>IDAS:</b> Índice de Desempenho de Adequação de Contratos, que avalia se os
contratos da operadora estão em conformidade com as regulamentações da ANS.

<b>IDGA:</b> Índice de Desempenho de Gestão de Processos e Regulação, que mede a
eficácia da gestão da operadora.

Esses indicadores são medidos em uma escala de 0 a 1, com 1 sendo o melhor
possível. São calculados com base em uma série de medidas específicas, que
variam dependendo do indicador. Eles são usados pela ANS para monitorar a
qualidade e a conformidade das operadoras de planos de saúde e fornecer aos
consumidores informações úteis para tomar decisões informadas sobre os planos de
saúde.

## 4. Processamento/Tratamento de Dados

<h3>Para realizar o processamento e o tratamento dos dados foi utilizada a linguagem de
programação python, uma vez que se torna mais possível e versátil ingerir e realizar
tratamentos desses dados selecionados. Com alguns problemas iniciais de
encoding, foi necessário importar no padrão: “ISO-8859-1” o recomendável é que
esses dados sempre estejam em “UTF-8”, utilizando a biblioteca Pandas os dados
foram carregados para um Dataframe para análises preliminares. E foram realizados
os seguintes tratamentos: 
    
- Renomear a coluna 'RazÆo Social' para 'RazaoSocial'.
- Adicionar zeros à esquerda e transformar a coluna para o formato string, para os valores relacionados a coluna “Registro” uma vez que se trata do código da operadora, não é um campo calculável. 
- Tratamento de encoding dos caracteres do ISO-558-9 para UTF-8

```python
# Importação da biblioteca
import pandas as pd
import numpy as np
import csv
```

```python
# Load the dataset
df = pd.read_csv(r'../Projeto AI/Bases/historico_idss.csv', delimiter=';', encoding='ISO-8859-1')

# Display the first few rows
df.info()
```

```python
# Visualizando as primeiras linhas do DataFrame
df.head()
```

```python
# Visualizando as últimas linhas do DataFrame
df.tail()
```

```python
# Renomear a coluna 'RazÆo Social' para 'RazaoSocial'
df.rename(columns={'RazÆo Social': 'RazaoSocial'}, inplace=True)
```

```python
# Converter valores da coluna 'Registro' para string
df['Registro'] = df['Registro'].astype(str)

# Preencher os valores com zeros à esquerda até formar 6 caracteres
df['Registro'] = df['Registro'].str.replace('\.0$', '').str.zfill(8)
df['Registro'] = df['Registro'].str.replace('.0', '') 
```

```python
# Mapeamento de caracteres problemáticos para correção ISO-885-9 para UTF-8 Campo: RazaoSocial
mappings = {
    'é': 'Ú',
    '�': 'É',
    '€' : 'Ç',
    'à' : 'Ó',
    'Ò' : 'Ê',
    'å' : 'Õ',
    'µ' : 'Á',
    'Ö' : 'Í',
    '¶' : 'Â',
    'Ç' : 'Ã',
    'â' : 'Ô',
    '' : 'É',
    'Ã': 'ÇÃ',
    'Õ': 'ÇÕ'
}

for original_char, correct_char in mappings.items():
    df['RazaoSocial'] = df['RazaoSocial'].str.replace(original_char, correct_char)

    
# Mapeamento de caracteres problemáticos para correção ISO-885-9 para UTF-8 Campo: Modalidade
mappings_2 = {
    '£': 'ú',
    '¢': 'ó',
    'Æ' : 'ã',
    'd' : 'éd'
}

for original_char, correct_char in mappings_2.items():
    df['Modalidade'] = df['Modalidade'].str.replace(original_char, correct_char)    
    
# (Opcional) Salvar o DataFrame corrigido como um novo arquivo CSV em UTF-8
df.to_csv(r'../Projeto AI/Bases/historico_idss_utf8.csv', index=False, encoding='utf-8-sig', sep=';')
```

## Identificando os valores faltantes

<h3>Os dados de todas as colunas são do tipo 'object', o que geralmente indica que são strings. Precisamos converter esses valores em números para a modelagem. Além disso, parece que temos 15 valores ausentes em várias colunas. Precisamos decidir como lidar com esses valores ausentes - podemos excluí-los ou preenchê-los com algum valor (como a média ou mediana, por exemplo).

Outro ponto a considerar é que algumas células contêm 'ND' (presumo que isso signifique 'Não Disponível' ou algo semelhante). Esses valores também precisarão ser tratados como valores ausentes.

Por fim, como mencionei antes, será útil reformatar os dados para que tenhamos uma coluna para o ano, uma para o tipo de indicador (IDSS, IDGR, etc.) e uma para o valor.

Vamos fazer essas transformações passo a passo. Primeiro, vou substituir 'ND' por NaN, e então converter as colunas numéricas de strings para números. Depois, lidaremos com os valores ausentes e reformataremos o dataframe.

<h3>

- Aplicação de "Melting": Usar a função melt do Pandas para reformatar o DataFrame.
- Redefinição das Colunas: Renomear as novas colunas para 'Ano', 'Indicador' e 'Valor'.

```python
# Separate identifier columns and indicator columns
identifier_columns = ['Registro', 'CNPJ', 'RazaoSocial', 'Modalidade']
indicator_columns = [col for col in df_cleaned.columns if col not in identifier_columns]

# Use the Pandas melt function to reshape the DataFrame
df_melted = pd.melt(df_cleaned, id_vars=identifier_columns, value_vars=indicator_columns, var_name='Indicador_Ano', value_name='Valor')

# Extract the year and indicator type from the 'Indicador_Ano' column
df_melted['Ano'] = df_melted['Indicador_Ano'].str.extract('(\d+)').astype('float')
df_melted['Indicador'] = df_melted['Indicador_Ano'].str.extract('([a-zA-Z]+)')

# Drop the original 'Indicador_Ano' column as we have separated it into two new columns
df_melted.drop(columns=['Indicador_Ano'], inplace=True)

# Show the first few rows of the reshaped DataFrame to validate the transformation
df_melted.head()
```

## 7. Aplicação de Modelos de Aprendizado de Máquina

Para realizar a modelagem preditiva de séries temporais, vou seguir os seguintes passos:
<h3>
    
- Seleção de Indicador: Vou selecionar um indicador para o qual o modelo de séries temporais será construído.
- Preparação de Dados: Os dados serão divididos em conjuntos de treinamento e teste.
- Seleção de Algoritmos: Selecionarei pelo menos 4 algoritmos de modelagem de séries temporais. 

  Os candidatos incluem:
  - ARIMA (Modelo AutoRegressivo Integrado de Médias Móveis);
  - SARIMA (ARIMA Sazonal);
  - Linear Regression;
  - LSTM;
- Treinamento e Teste: Cada modelo será treinado e testado.
- Avaliação de Desempenho: Mean Squared Error (MSE): e Mean Absolute Error (MAE)

```python
# Importação de bibliotecas
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np
from sklearn.linear_model import LinearRegression
```

### Seleção de Indicador

```python
# Extract time series data for IDSS
time_series_data = df_melted[df_melted['Indicador'] == 'IDSS'].groupby('Ano')['Valor'].mean().reset_index()
time_series_data.columns = ['Ano', 'Valor']
```

### Separação de Treino e Teste (70-30% split)

```python
# Split data into training and test sets (70-30% split)
train_size = int(len(time_series_data) * 0.7)
train, test = time_series_data.iloc[:train_size], time_series_data.iloc[train_size:]
```

### Inicializando métricas

```python
# Initialize metrics dictionary
metrics = {'Model': [], 'Mean Squared Error': [], 'Mean Absolute Error': []}
```

### Treinamento do Modelo: ARIMA

```python
# ARIMA model
try:
    arima_model = ARIMA(train['Valor'], order=(1, 1, 1))
    arima_fit = arima_model.fit()
    arima_forecast = arima_fit.forecast(steps=len(test))
    arima_mse = mean_squared_error(test['Valor'], arima_forecast)
    arima_mae = mean_absolute_error(test['Valor'], arima_forecast)
    metrics['Model'].append('ARIMA')
    metrics['Mean Squared Error'].append(arima_mse)
    metrics['Mean Absolute Error'].append(arima_mae)
except Exception as e:
    arima_mse, arima_mae = np.nan, np.nan
```

### Treinamento do Modelo: SARIMA

```python
# SARIMA model
try:
    sarima_model = SARIMAX(train['Valor'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    sarima_fit = sarima_model.fit()
    sarima_forecast = sarima_fit.forecast(steps=len(test))
    sarima_mse = mean_squared_error(test['Valor'], sarima_forecast)
    sarima_mae = mean_absolute_error(test['Valor'], sarima_forecast)
    metrics['Model'].append('SARIMA')
    metrics['Mean Squared Error'].append(sarima_mse)
    metrics['Mean Absolute Error'].append(sarima_mae)
except Exception as e:
    sarima_mse, sarima_mae = np.nan, np.nan
```

### Treinamento do Modelo: Linear Regression

```python
# Linear Regression model
try:
    lr_model = LinearRegression()
    lr_model.fit(train[['Ano']], train['Valor'])
    lr_forecast = lr_model.predict(test[['Ano']])
    lr_mse = mean_squared_error(test['Valor'], lr_forecast)
    lr_mae = mean_absolute_error(test['Valor'], lr_forecast)
    metrics['Model'].append('Linear Regression')
    metrics['Mean Squared Error'].append(lr_mse)
    metrics['Mean Absolute Error'].append(lr_mae)
except Exception as e:
    lr_mse, lr_mae = np.nan, np.nan
```

### Treinamento do Modelo: LSTM

```python
# Prepare data for LSTM
X = train['Valor'].values.reshape(-1, 1, 1)
y = train['Valor'].values
X_test = test['Valor'].values.reshape(-1, 1, 1)
y_test = test['Valor'].values

# LSTM model
try:
    lstm_model = Sequential()
    lstm_model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
    lstm_model.add(Dense(1))
    lstm_model.compile(optimizer='adam', loss='mse')
    lstm_model.fit(X, y, epochs=300, verbose=0)
    lstm_forecast = lstm_model.predict(X_test)
    lstm_mse = mean_squared_error(y_test, lstm_forecast)
    lstm_mae = mean_absolute_error(y_test, lstm_forecast)
    metrics['Model'].append('LSTM')
    metrics['Mean Squared Error'].append(lstm_mse)
    metrics['Mean Absolute Error'].append(lstm_mae)
except Exception as e:
    lstm_mse, lstm_mae = np.nan, np.nan
```
